```python
y_train = y_train/100
```

```python
# import XGB regression model
from xgboost import XGBRegressor


# declare parameters
# params = {
#             learning_rate': 0.10803312273087397, 'max_depth': 3, 'n_estimators': 103, 'subsample': 0.813623786492104
#         }
            
# instantiate the regression model (**params)
xgbr = xgb.XGBRegressor(verbosity=0
                        , objective='reg:squarederror'
                        #, n_estimators=100
                        #,num_round=2000
                        , learning_rate=0.3
                        #, max_depth=5
                        #, subsample=0.9
                        #, colsample_bytree=0.5
                        #, gamma=1
                        #,min_child_weight=0.01
                        #, gamma=0.0
                        ) 



# fit the classifier to the training data+
xgbr.fit(X_train, y_train)
```

```python
# Create the feature importance plot
xgb.plot_importance(xgbr)

# Adjust the figure size to provide more space for labels
plt.rcParams['figure.figsize'] = [8, 24]

# Save the plot as 'feature_importance.png'
plt.savefig('feature_importance.png')

# Display the plot
plt.show()
```

```python
score = xgbr.score(X_train, y_train)  
print("Training score: ", score)
```

```python
from sklearn.metrics import mean_squared_error

# make predictions on test data
y_pred = xgbr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("MSE: %.2f" % mse)
```
